# Tentative d'une représentaion heatmap de données aero

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sb
```

```python
df = pd.read_excel("carto_aero.xlsx")
df.columns =  ["x","y","v"]
```

```python
import matplotlib
matplotlib.style.use('ggplot')
```

```python
# n = 5000
df2 = df 
# df2 = df.sample(n)
ax = df2.plot.scatter(figsize=(10, 7),x='x',y='y',c='v',
                 grid=False,colormap="viridis")
ax.set_ylabel('échelle');
ax.set_xlabel('x');
ax.set_title("HeatMap de carto_aero");
```

```python
df2 = df 

ax = df2.plot.scatter(figsize=(10, 7),x='x',y='y',c='v',xlim=[-50,50],ylim=[-50,50],
                 grid=False,colormap="viridis")
ax.set_ylabel('échelle');
ax.set_xlabel('x');
ax.set_title("HeatMap de carto_aero Zoom 50");
```
